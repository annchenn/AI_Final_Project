```python
import os
import json
import time
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
```

```python
options = Options()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

all_recipes = []
data_count = 0

for i in range(1, 6):
    url = f"https://icook.tw/recipes/popular?page={i}"
    driver.get(url)
    time.sleep(3)

    cuisine_items = driver.find_elements(By.CSS_SELECTOR, 'li.browse-recipe-item')

    for item in cuisine_items:
        try:
            link_element = item.find_element(By.CSS_SELECTOR, 'a.browse-recipe-link')
            name = link_element.get_attribute('aria-label')
            url = link_element.get_attribute('href')
            all_recipes.append({"count": data_count+1, "name": name, "url": url})
            data_count+=1
        except Exception as e:
            print("can't get cuisine items")

with open("recipe_urls.json", "w") as f:
    json.dump(all_recipes, f, indent=2)
```

```python
with open("recipe_urls.json", "r") as f:
    cuisine_items = json.load(f)

all_cuisine_data = []

for cuisine in cuisine_items:
    url = cuisine["url"]

    driver.get(url)
    time.sleep(3)

    cuisine_dic = {}
    # 下載圖片
    image_dir = 'images'
    cuisine_id = url[-6:]
    cuisine_dic["cuisine_id"] = cuisine_id

    try:
        cover_img = driver.find_element(By.CSS_SELECTOR, 'div.recipe-cover img')
        cover_img_url = cover_img.get_attribute('src')
        cuisine_dic["cuisine_img_url"] = cover_img_url
        # cover_img_data = requests.get(cover_img_url).content
        # cover_img_path = os.path.join(image_dir, f"{cuisine_id}.jpg")
        # with open(cover_img_path, 'wb') as f:
        #     f.write(cover_img_data)

        # print(f"Saved cuisine image as {cuisine_id}.jpg")

    except Exception as e:
        print(f"Can't fine the cuisine image：{e}")
        cover_img_url = ""
        cover_img_path = ""


    try:
        cuisine_name = driver.title
        if "by" in cuisine_name:
            cuisine_name = cuisine_name.split("by")[0].strip()
        else:
            cuisine_name = cuisine_name.strip()
        cuisine_dic["cuisine_name"] = cuisine_name
    except:
        cuisine_dic["cuisine_name"] = ""


    # 列出原料
    cuisine_dic["ingredients"]=[]
    ingredients_item =driver.find_elements(By.CSS_SELECTOR, 'li.ingredient')
    for ingredient in ingredients_item:
        ingredient_name = ingredient.find_element(By.CSS_SELECTOR, 'a.ingredient-search')
        ingredient_name = ingredient_name.get_attribute('data-name')
        cuisine_dic["ingredients"].append(ingredient_name)
        print(ingredient_name)

    # 找到所有步驟 li 元素
    cuisine_dic["cooking_instructions"]=""
    steps = driver.find_elements(By.CSS_SELECTOR, 'li.recipe-details-step-item')

    for idx, step in enumerate(steps, start=1):
        try:
            # cooking steps
            desc = step.find_element(By.CSS_SELECTOR, 'p.recipe-step-description-content').text
            cuisine_dic["cooking_instructions"] = cuisine_dic["cooking_instructions"]+desc+"; "
            
            print(f"Step {idx}")
            print("描述：", desc)
            print("-" * 40)

        except Exception as e:
            print(f"Step {idx} 發生錯誤：{e}")
            
    all_cuisine_data.append(cuisine_dic)


driver.quit()


with open("all_cuisine_recipe.json", "w", encoding="utf-8") as f:
    json.dump(all_cuisine_data, f, ensure_ascii=False, indent=2)

print("finished!")
```

```python

options = Options()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
url = "https://icook.tw/recipes/477023"
driver.get(url)
time.sleep(3)

cuisine_dic = {}
# 下載圖片
image_dir = 'images'
cuisine_id = 404891
cuisine_dic["id"] = cuisine_id

try:
    cover_img = driver.find_element(By.CSS_SELECTOR, 'div.recipe-cover img')
    cover_img_url = cover_img.get_attribute('src')
    cuisine_dic["img_url"] = cover_img_url
    cover_img_data = requests.get(cover_img_url).content
    cover_img_path = os.path.join(image_dir, f"{cuisine_id}.jpg")
    with open(cover_img_path, 'wb') as f:
        f.write(cover_img_data)

    print(f"Saved cuisine image as {cuisine_id}.jpg")

except Exception as e:
    print(f"Can't fine the cuisine image：{e}")
    cover_img_url = ""
    cover_img_path = ""


# 找到所有步驟 li 元素

steps = driver.find_elements(By.CSS_SELECTOR, 'li.recipe-details-step-item')

for idx, step in enumerate(steps, start=1):
    try:
        # cooking steps
        desc = step.find_element(By.CSS_SELECTOR, 'p.recipe-step-description-content').text
        cuisine_dic["cooking_instructions"] = cuisine_dic["cooking_instructions"]+desc
        print(f"Step {idx}")
        print("描述：", desc)
        print("-" * 40)

    except Exception as e:
        print(f"Step {idx} 發生錯誤：{e}")


driver.quit()


with open("all_steps.json", "w", encoding="utf-8") as f:
    json.dump(cuisine_dic, f, ensure_ascii=False, indent=2)

print("finished!")
```
